```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('housing.csv', header=None, delimiter=r'\s+')
y = data[13]
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression 

X = data.iloc[:,0:2]
lr = LinearRegression() 
lr.fit(X, y)

y_pred = lr.predict(X)

mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("2 признака")
print(f"MAE: {mae}")
print(f"R2: {r2}")
print(f"MSE: {mse}")
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x1 = X.iloc[:, 0]
x2 = X.iloc[:, 1]

ax.scatter(x1, x2, y, color='red')

x1_range = np.linspace(x1.min(), x1.max(), 20)
x2_range = np.linspace(x2.min(), x2.max(), 20)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

x_grid_flat = np.c_[x1_grid.ravel(), x2_grid.ravel()]
y_grid_pred = lr.predict(x_grid_flat).reshape(x1_grid.shape)

ax.plot_surface(x1_grid, x2_grid, y_grid_pred, alpha=0.6)

ax.set_xlabel('Первый признак')
ax.set_ylabel('Второй признак')

plt.show()
```

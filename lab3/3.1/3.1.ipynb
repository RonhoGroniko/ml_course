```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df
```

```python
print("Названия сортов:", iris.target_names)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for t in np.unique(df['target']):
    plt.scatter(df[df['target'] == t]['sepal length (cm)'], df[df['target'] == t]['sepal width (cm)'], label=iris.target_names[t])
plt.title('Sepal')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.legend()

plt.subplot(1, 2, 2)
for t in np.unique(df['target']):
    plt.scatter(df[df['target'] == t]['petal length (cm)'], df[df['target'] == t]['petal width (cm)'], label=iris.target_names[t])
plt.title('Petal')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='target', palette='bright')
```

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='target', palette='bright')
```

```python
df1 = df[df['target'].isin([0, 1])]
df2 = df[df['target'].isin([1, 2])]
```

```python
from sklearn.model_selection import train_test_split 
X1_train, X1_test, y1_train, y1_test = train_test_split(df1.iloc[:, :-1], df1['target'], test_size=0.3, stratify=df1["target"])
X2_train, X2_test, y2_train, y2_test = train_test_split(df2.iloc[:, :-1], df2['target'], test_size=0.3, stratify=df2["target"])
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
```

```python
from sklearn.metrics import accuracy_score
clf.fit(X1_train, y1_train)
y1_pred = clf.predict(X1_test)
print("Точность:", accuracy_score(y1_test, y1_pred))
```

```python
clf.fit(X2_train, y2_train)
y2_pred = clf.predict(X2_test)
print("Точность:", accuracy_score(y2_test, y2_pred))
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', alpha=0.7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Точность:", accuracy_score(y_test, y_pred))
```
